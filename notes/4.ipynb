# Практическое занятие №4: Случайные сигналы, шумы и спектральная плотность мощности

## 1. Генерация белого, розового и красного шума

1.  **Гауссовый белый шум:** `np.random.normal(0, scale, N)`.
2.  **Розовый шум:**
    - Выполнить БПФ белого шума (`np.fft.rfft`).
    - Создать массив частот `freqs = np.fft.rfftfreq(N, d=1/fs)`.
    - Домножить спектр на `1 / np.sqrt(freqs)`. **Важно:** избегать деления на ноль на нулевой частоте (установить множитель = 0 или 1).
    - Выполнить обратное БПФ (`np.fft.irfft`).
    - Нормализовать до единичной дисперсии (`signal / np.std(signal)`).
3.  **Красный шум:** `np.cumsum(white_noise)` и затем нормировка до единичной дисперси.

## 2. Генерация шума с разной дисперсией

- Для белого шума: `scale=np.sqrt(disp)`.
- Для красного и розового шума после генерации умножить на `np.sqrt(disp)`.

## 3. Оценка спектральной плотности мощности (СПМ)

- Вычисление периодограммы: `f_per, Pxx_per = scipy.signal.periodogram(signal, fs)`
- Вычисление СПМ методом Уэлча: `f_welch, Pxx_welch = scipy.signal.welch(signal, fs, nperseg=256, noverlap=128, window='hann')`
- Построение графиков в двойном логарифмическом масштабе: `plt.loglog(f, Pxx)`:
если $y=x^p$, то $\lg y = p\lg x$.

### 4. Отношение сигнал-шум (SNR)

**Расчет амплитуды шума по SNR:**
- Мощность синусоиды: `P_sig = A² / 2`.
- SNR (линейный) = `10^(SNR_dB/10)`.
- Мощность шума: `P_noise = P_sig / SNR_lin`.
- Дисперсия шума = `P_noise`.
- `noise = np.random.normal(0, np.sqrt(P_noise), N)`.
